# Data Cleaning

```python
# it returns how many null values in a particular column
df.isnull().sum()
```

```python
df['Income']
```

```python
mean=round(df['Income'].mean())
mean
```

```python
# replacing null values into mean value
df_new = df.fillna({'Income':mean})
df_new
```

```python
# re-checking
df_new.isnull().sum()
```

```python
df_new
```

```python
df_new['Dt_Customer'] = pd.to_datetime(df_new['Dt_Customer'],format='%d-%m-%Y',errors='coerce')
df_new
```
